## Navegating files

```python
import os
```

```python
directory = 'C:/Users/Daniel/Desktop/Capsulas/Analise/Experimento Capsulas 2°/Imagens'
rows = []
for tempo in os.listdir(directory):
    print("- " + tempo)
    for composicao in os.listdir(directory + "/" + tempo):
        print("-- " + composicao)
        if tempo == "T0":
            for imagem in os.listdir(directory + "/" + tempo + "/" + composicao):
                print("--- " + imagem)
                zoom = int(imagem.split("X_")[0].split("_")[-1])
                if zoom == 4:
                    endereco = tempo + "/" + composicao + "/" + imagem
                    rows.append([tempo, composicao, '-', endereco, []])
        else:
            for condicao in os.listdir(directory + "/" + tempo + "/" + composicao):
                print("--- " + condicao)
                for imagem in os.listdir(directory + "/" + tempo + "/" + composicao + "/" + condicao):
                    print("---- " + imagem)
                    zoom = int(imagem.split("X_")[0].split("_")[-1])
                    if zoom == 4:
                        endereco = tempo + "/" + composicao + "/" + condicao + "/" + imagem 
                        rows.append([tempo, composicao, condicao, endereco, []])
```

## Creating dataframe

```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
columns = ['tempo', 'composicao', 'condicao', 'endereco', 'dados']
df = pd.DataFrame(rows, columns = columns)
df
```

## Computer Vision

```python
import cv2 as cv     # visão computacional
import numpy as np   # manipulação matemática
from PIL import Image   # só pra mostrar a imagem
```

```python
for index, row in df.iterrows():
    path = directory + "/" + row['endereco']
    params = {
        'distance': 8,
        'radius_tolerance': 53,
        'center_tolerance': 53,
        'min_radius': 80,
        'max_radius': 200
    }
    if row['condicao'] == 'osmotico':        
        params['min_radius'] = 120
        params['max_radius'] = 300

    
    # PREPROCESSING IMAGES
    src = np.asarray(Image.open(path))
    gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
    
    # CALCULATING CIRCLES
    rows = gray.shape[0]
    circles = cv.HoughCircles(gray, cv.HOUGH_GRADIENT, 1,
                              rows / params['distance'],
                              param1 = params['radius_tolerance'],
                              param2 = params['center_tolerance'],
                              minRadius = params['min_radius'],
                              maxRadius = params['max_radius'])

    # SAVING RESULTS
    radius = []
    detection = src.copy()
    if circles is not None:
        circlesInt = np.uint16(circles)[0]
        for x, y, r in circlesInt:
            radius.append(r)
            cv.circle(detection, (x, y), 3, (255, 0, 0), 3)
            cv.circle(detection, (x, y), r, (0, 255, 0), 3)
            cv.putText(detection, str(np.round(r)), (x + 10, y), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv.LINE_AA)


    df.at[index, 'dados'] = radius

    path_proc = "C:/Users/Daniel/Desktop/Capsulas/Analise/Experimento Capsulas 2°/Imagens Processadas"
    image_name_src = str(index) + '.jpeg'
    image_name_debug = str(index) + '_.jpeg'

    Image.fromarray(src).save(path_proc + "/" + image_name_src)
    Image.fromarray(detection).save(path_proc + "/" + image_name_debug)
    
    
    print('done with ', index)
```

## STATISTICS
